Memadukan fungsi agregasi dengan transform()

```python
import pandas as pd

print(pd.__version__)
```

```python
d ={'no_nota' : [1,1,1,2,2,3,4,5],
    'kopi' : ['latte', 'cappuccino', 'expresso', 'latte', 'expresso', 'cappuccino', 'latte','expresso'],
    'harga' : [50,60,80,150,120,60,100,40]}

df = pd.DataFrame(d)
df
```

Menghitung total harga untuk tiap nomor nota

```python
df.groupby('no_nota')['harga'].sum().to_frame()
```

```python
df['total_harga'] = df.groupby('no_nota')['harga'].transform(sum)
df
# Membuat no_nota yang bersangkutan memiliki total_harga keseluruhan no_nota
```

Menghitung total omset untuk tiap jenis kopi yang terjual

```python
df.groupby('kopi')['total_harga'].sum().to_frame()
```

```python
df['total_omset'] = df.groupby('kopi')['total_harga'].transform(sum)
df
# Karena dataframe harus memiliki 8 row sedangkan sum hanya ada 3 row, dapat digunakan transform
```
